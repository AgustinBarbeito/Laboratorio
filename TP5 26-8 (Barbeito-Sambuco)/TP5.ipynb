```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as grafico
df = pd.read_csv("StudentsPerformance.csv")
```

Cual es el promedio de notas en escritura en relacion al nivel educativo parental?

```python
promedio = df.groupby('parental level of education')['writing score'].mean().sort_values(ascending=False)

x = promedio.index
y = promedio

grafico.bar(x,y)
grafico.xlabel('Nivel Educativo Parental')
grafico.ylabel('Promedio')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()

```

Podemos ver que el nivel educativo parental no tiene relacion directa con las notas, al menos de escritura. Casi todos presentan las mismas notas.

Cual es la cantidad de que cada nivel educativo parental?

```python
ordenado = df.sort_values(by=['parental level of education'], ascending=False)

x = ordenado["parental level of education"].unique()
y = ordenado["parental level of education"].value_counts()

grafico.bar(x,y)
grafico.xlabel('Nivel Educativo Parental')
grafico.ylabel('Cantidad')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```

Viendo el grafico podemos ver que hay bastantes con "Some High School y Some College" y por el otro lado vemos que hay pocos con un "Bachelor o Associate Degree". Esto es entendible, al ser estos ultimos dos de nivel mas avanzado que los primeros dos.

Cual es la cantidad de cada nivel de grupos etnicos?

```python
ordenado = df.sort_values(by=['race/ethnicity'], ascending=False)

x = ordenado["race/ethnicity"].unique()
y = ordenado["race/ethnicity"].value_counts()

grafico.bar(x,y)
grafico.xlabel('Grupo Etnico')
grafico.ylabel('Cantidad')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```

Podemos ver una gran cantidad entre el grupo E y D, practicamente duplicando o mas al grupo A.

Que genero presenta el mejor promedio de notas en matematica?

```python
promedio = df.groupby('gender')['math score'].mean().sort_values(ascending=False)

x = promedio.index
y = promedio

grafico.bar(x,y)
grafico.xlabel('Genero')
grafico.ylabel('Notas de Matematica')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```

Con el grafico podemos ver que a pesar de que en promedio los hombres tienen mejores notas que las mujeres en matematica, los resultados son bastante parejos entre los 2 generos.

Cual es el porcentaje de alumnos que completaron el curso de preparacion?

```python
preparacion= df["test preparation course"].unique()

cantidad= df["test preparation course"].value_counts()
grafico.pie(cantidad, labels=preparacion, autopct='%1.1f%%')
grafico.show()
```

Viendo el grafico de torta, vemos que casi 2/3 del curso no realizo el curso de preparacion. Es bastante impactante ese numero sabiendo que todos tuvieron la oportunidad de realizarlo.
