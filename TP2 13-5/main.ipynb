```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2
data=pd.read_csv("../data/listado-de-personal-2010-05-v2.csv")
data=data.dropna()
data=data.loc[0:150]
```

```python
x=data['Gerencia'].unique()
y=data['Gerencia'].value_counts().tolist()
gf1.bar(x,y)
gf1.title('TipoDeGerencia')
gf1.xlabel('Gerencia')
gf1.ylabel('CantidadDeGerencias')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()
```

```python
x2=data['Puesto'].unique()
y2=data['Puesto'].value_counts().tolist()
gf2.bar(x2,y2)
gf2.title('Puestos')
gf2.xlabel('Puesto')
gf2.ylabel('CantidadDePuestos')
gf2.xticks(rotation=90,fontsize=12)
gf2.show()
```
