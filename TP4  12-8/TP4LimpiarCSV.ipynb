```python
import numpy as np
import pandas as pd
import matplotlib as plt

def EliminarMinusculas(valor):
  if type(valor) == str:
    return valor.lower()
  else:
    return valor

def SacarEspacios(valor):
    if type(valor) == str:
        return valor.replace(" ", "")
    else:
        return valor

def Nan(valor):
    if type(valor) == str:
        if valor == "nan" or valor == "" or valor == " ":
            return np.nan
    return valor

def ArreglarAltura(valor):
    valor = str(valor)
    return valor.replace("m","")

def ArreglarPeso(valor):
    valor = str(valor)
    for character in valor:
        if not character.isnumeric():
            return valor.split(character, 1)[0]
    return valor

def ArreglarSexo(valor):
    if "mu" in valor:
        return "m"
    if "ho" in valor:
        return "h"
    if "binari" in valor:
        return "nb"
    if "h" in valor or "m" in valor or "nb" in valor:
        return valor
    return np.nan

def Observaciones(valor):
    valor = str(valor)
    if "hiper" in valor and "tens" in valor:
        return "hipertension"
    if "dolor" in valor and "cabeza" in valor:
        return "dolor de cabeza"
    return "ninguna"

def ArreglarLu(valor):
    valor = str(valor)
    valor_list = list(valor)
    if valor[len(valor)-3] != "/" :
        if valor[len(valor)-3].isnumeric():
            valor_list.insert(len(valor_list)-2, "/")
        else:
            valor_list[len(valor_list)-3] = "/"
    elif valor[len(valor)-4] == "/" :
        valor_list[len(valor)-4] = ""
    return "".join(valor_list)



df = pd.read_csv("data/datos.csv")
df = df.applymap(SacarEspacios)
df = df.applymap(EliminarMinusculas)
df[" Altura"] = df[" Altura"].apply(ArreglarAltura)
df[" Peso"] = df[" Peso"].apply(ArreglarPeso)
df[" Sexo"] = df[" Sexo"].apply(ArreglarSexo)
df = df.applymap(Nan)
df.dropna(inplace=True)
df[" LU"] = df[" LU"].apply(ArreglarLu)
df[" Observacion"] = df[" Observacion"].apply(Observaciones)
print(df)
```
