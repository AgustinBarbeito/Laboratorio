```python
import pandas as pd
import matplotlib.pyplot as gf1


data = pd.read_csv("../data/formulario.csv")
```

```python
x = data['Meses'].unique()
y = data['Meses'].value_counts().tolist()

y.reverse()
gf1.bar(x,y)
gf1.title('Meses con mas nacimientos')
gf1.xlabel('Meses')
gf1.ylabel('Cantidad de nacimientos')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()


```

Introduccion:

-Mi encuesta es sobre los meses con mayor cantidad de nacimientos. Los meses estaran separados en 6 pares.
La encuesta se realizo a 25 personas y los medios de comunicacion utilizados fueron Whatsapp y Discord.

Conclusion:

-Las conclusion obtenidas fueron que hay una gran tasa de nacimientos en los entrevistados en los meses de Noviembre-Diciembre y Septiembre-Octubre, practicamente sobre el ultimo cuarto del a√±o. Al contrario, en Julio-Agosto se presento la menor cantidad de casos.
