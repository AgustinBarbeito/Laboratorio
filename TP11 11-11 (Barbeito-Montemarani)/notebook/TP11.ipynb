```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('..\data\Analisis taller de soporte informático (respuestas) - Respuestas de formulario 1.csv')
del df["Marca temporal"]
```

# ¿Cuáles son las razones por las que el alumno considera que no ha entregado todos los trabajos?

¿Tendrá que ver con el lenguaje de programación?¿Cuántos odian python?

```python
X=df["¿Cuánto te gusta Python como lenguaje de programación?"].unique()
Y=df["¿Cuánto te gusta Python como lenguaje de programación?"].value_counts().tolist()

plt.bar(X,Y)
plt.title('¿Cuánto te gusta Python como lenguaje de programación?')
plt.xlabel('Respuestas')
plt.ylabel('Cantidad')
plt.show()
```

Segun este grafico el problema no esta en el lenguaje ya que son muy pocos los que no les gusta python.

¿Tendrá que ver con el horario?¿Cuántos consideran que su rendimiento en las primeras horas de la mañana es menor que en la tarde?

```python
label=df["¿Consideras que tu rendimiento por la mañana es peor que por la tarde?"].unique()
cantidad=df["¿Consideras que tu rendimiento por la mañana es peor que por la tarde?"].value_counts().tolist()

plt.pie(cantidad,labels = label, autopct='%1.1f%%')
plt.title('¿Consideras que tu rendimiento por la mañana es peor que por la tarde?')
plt.show()
```

Como se ve en el grafico esta podria ser una de las principales causas, la mitad de los alumnos consideran que trabajan mejor por la tarde que por la mañana.

¿Tendrá que ver con los contenidos?¿Cuántos consideran que analizar datos es aburrido?

```python
label=df["¿Considerás que analizar datos es aburrido?"].unique()
cantidad=df["¿Considerás que analizar datos es aburrido?"].value_counts().tolist()

plt.pie(cantidad,labels = label, autopct='%1.1f%%')
plt.title('¿Considerás que analizar datos es aburrido?')
plt.show()
```

Esta podria ser otra de las causas ya que casi el 40% de los alumnos lo consideran aburrido.

¿Hasta qué mes del año se han entregado la mayoría los trabajos?

```python
X=df["¿Hasta qué mes entregaste trabajos?"].unique()
Y=df["¿Hasta qué mes entregaste trabajos?"].value_counts().tolist()

plt.bar(X,Y)
plt.title('¿Hasta qué mes entregaste trabajos?')
plt.xlabel('Respuestas')
plt.ylabel('Cantidad')
plt.show()
```

La mayoria entrego trabajos hasta agosto, puede ser porque llegando a fin de año empezamos a apurarnos mas con otras materias como practicas profesionalizantes.
