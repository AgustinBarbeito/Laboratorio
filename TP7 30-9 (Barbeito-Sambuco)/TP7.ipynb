```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("data\StudentsPerformance.csv")

df1 = df[["parental level of education","math score", "writing score"]]
grafico = sns.pairplot(df1,hue="parental level of education",palette="Spectral")
```

Conclusion: No hay relacion entre el Nivel Educativo Parental y las notas de matematica y escritura de sus hijos, pero si hay una relacion proporcional entre las notas. Los que tienen mejores notas en matematica tambien las tienen en escritura.
